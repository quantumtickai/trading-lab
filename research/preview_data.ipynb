```python
qlib.init(provider_uri="X:/QuantumTickAI/trading-lab/data/raw/us_data", region=REG_US) # Adjust the path as needed
print("✅ QLib initialized.")
```

```python
instrument = "AAPL"
```

```python
df = D.features([instrument], ["$open", "$high", "$low", "$close", "$volume"],
                start_time="2011-03-07", end_time="2011-07-25")
```

```python
print(df.head())
```

```python
# 4. Plot Closing Price
df["$close"].plot(figsize=(12,6), title=f"{instrument} Closing Price")
plt.show()
```
